```python
import pandas as pd
import numpy as np
import requests
from io import BytesIO

%load_ext autoreload
%autoreload 2

import warnings
warnings.filterwarnings("ignore")
```

```python
url = 'https://www.estadisticaciudad.gob.ar/eyc/wp-content/uploads/2025/05/PBP_CO_1020.xls'
ruta_archivo = 'Datasets/Evolucion_poblacion_CABA.xls'

response = requests.get(url)

if response.status_code == 200:
    excel_content = BytesIO(response.content)

    # Leer el archivo Excel
    df = pd.read_excel(excel_content)

    # Ahora 'df' contiene los datos del archivo Excel
    print(df.head())

else:
    print(f'Error al realizar la solicitud. Código de estado: {response.status_code}')
```

```python
df #vemos que hay mucha informacion de mas
```

```python
columnas = ['Comuna']
for i in range(2010,2026):
    columnas.append(i)
 # definimos nombre de columnas
columnas
```

```python
df.columns= columnas
df = df.iloc[3:18].reset_index(drop=True)
df #quedemonos con lo que nos interesa, ambos sexos
```

```python
df.drop(columns=[2010,2011,2012,2013,2014,2015,2022,2023,2024,2025],inplace=True) #elijamos las columnas
df
```

Bueno, estas son las proyecciones realizadas por la ciudad de buenos aires y a los efectos de nuestro analisis las vamos a tomar como validas, mas alla de que pueden discurrir de la realidad exacta.

```python
# pasemos este analisis a un formato de base de datos:

df_melted = pd.melt(df, id_vars=['Comuna'], var_name='Anio', value_name='Habitantes')

df_melted_total_anio = df_melted.groupby('Anio', as_index=False)['Habitantes'].sum()
df_melted_total_anio
df_melted_con_totales = pd.concat([df_melted, df_melted_total_anio]).reset_index(drop=True)
df_melted_con_totales
```
